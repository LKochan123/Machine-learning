```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np

# load the famous iris data
irisRaw = load_iris()
 
#Convert data to Panda package
# read iris.data into a pandas DataFrame (df), including column names
iris = pd.DataFrame(data= np.c_[irisRaw['data'], irisRaw['target']],
                     columns= irisRaw['feature_names'] + ['target'])
```

#### Zad 8)

```python
kk = 'petal length (cm)'
max_petal_length, max_idx = max(iris[kk]), iris[kk].idxmax()
min_petal_length, min_idx = min(iris[kk]), iris[kk].idxmin()

print(f'max_value = {max_petal_length}, index = {max_idx}')
print(f'min_value = {min_petal_length}, index = {min_idx}')
```

#### Zad 9)

```python
iris_col_names = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)', 'target']

for i in range(len(iris_col_names)):
    std = iris[iris_col_names[i]].std()
    print(f'Odchylenie kolumny nr {i} wynosi {std}.')
```

#### Zad 11)

```python
import matplotlib.pyplot as plt 

def show_plot_of_iris(col_name, color, bins=50):
    plt.hist(iris[col_name], color=color, bins=bins)
    plt.xlabel(f'{col_name}')
    plt.ylabel('count')
    plt.title(f'{col_name[0].upper()}{col_name[1:-5]} of iris')
    plt.show()

show_plot_of_iris('sepal width (cm)', 'green')
show_plot_of_iris('sepal length (cm)', 'orange', 80)
show_plot_of_iris('petal width (cm)', 'lightblue')
show_plot_of_iris('petal length (cm)', 'grey', 120)
```
